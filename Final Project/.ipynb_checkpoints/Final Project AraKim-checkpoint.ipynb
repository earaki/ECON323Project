The FIFA/Coca-Cola World Ranking is calculated using the Elo Model, which adjusts a team's point total based on match outcomes. The number of points added or deducted depends on several factors, including 1) The relative strength of the teams involved 2) The importance of the match 3) The actual match result 4) The expected match result.

**Number of professional players in Europe's top 5 leagues, grouped by nationality (2003â€“2023)**

The number of players represents the number of players from each country that play the the top 5 European leagues of football. This data was accessed through [fbref.com](https://fbref.com/en/comps/Big5/2010-2011/nations/2010-2011-Big-5-European-Leagues-Nationalities). 

```python
#List of countries to filter
filtered_teams = [
    "England", "Spain", "Germany", "Brazil", "Italy", "Argentina", "Mexico",
    "United States", "France", "Saudi Arabia", "TÃ¼rkiye", "Qatar", "Korea Republic",
    "Indonesia", "Colombia", "Portugal", "Netherlands", "Poland", "Scotland",
    "Switzerland"]

#Load player data from multiple CSV files, filter by the given countries, and merge the results.
def load_and_filter_players_data(file_paths, filtered_teams):

    dfs = []
    
    for file_path in file_paths:
        df = pd.read_csv(file_path)
        df['Nation'] = df['Nation'].apply(lambda x: ' '.join(x.split()[1:]))
        if '# Players' in df.columns:
            df = df[df['Nation'].isin(filtered_teams)]
            year = file_path.split('/')[-1].split('-')[0]
            df = df[['Nation', '# Players']]
            df = df.rename(columns={'# Players': f'{year}'})
            dfs.append(df)

    if dfs:
        merged_df = dfs[0]
        for df in dfs[1:]:
            merged_df = pd.merge(merged_df, df, on='Nation', how='outer')

        merged_df.set_index('Nation', inplace=True)
        return merged_df
    else:
        print("No dataframes to merge.")
        return pd.DataFrame()

# Example usage with the uploaded files
file_paths = ["Data sets/NumberPlayers/2003-2004.csv", "Data sets/NumberPlayers/2004-2005.csv", 
              "Data sets/NumberPlayers/2005-2006.csv", "Data sets/NumberPlayers/2006-2007.csv", 
              "Data sets/NumberPlayers/2007-2008.csv", "Data sets/NumberPlayers/2008-2009.csv", 
              "Data sets/NumberPlayers/2009-2010.csv", "Data sets/NumberPlayers/2010-2011.csv", 
              "Data sets/NumberPlayers/2011-2012.csv", "Data sets/NumberPlayers/2012-2013.csv", 
              "Data sets/NumberPlayers/2013-2014.csv", "Data sets/NumberPlayers/2014-2015.csv",
              "Data sets/NumberPlayers/2015-2016.csv", "Data sets/NumberPlayers/2016-2017.csv",
              "Data sets/NumberPlayers/2017-2018.csv", "Data sets/NumberPlayers/2018-2019.csv", 
              "Data sets/NumberPlayers/2019-2020.csv", "Data sets/NumberPlayers/2020-2021.csv", 
              "Data sets/NumberPlayers/2021-2022.csv", "Data sets/NumberPlayers/2022-2023.csv", 
              "Data sets/NumberPlayers/2023-2024.csv"]
players_df = load_and_filter_players_data(file_paths, filtered_teams)
players_df.reset_index(inplace=True)
players_df = players_df.rename(columns = {'Nation':'Country'})
players_df["Country"] = players_df["Country"].replace("TÃ¼rkiye", "Turkey") #ensure country names are same for merging
players_df["Country"] = players_df["Country"].replace("Korea Republic", "South Korea")
players_df.to_csv("players_df.csv", index=False)
players_df
```

```python
#Step 3: Merging the first two datasets into one
fifa_ranking_df_long = fifa_ranking_df.melt(id_vars=["Country"], var_name="Year", value_name="FIFA Ranking")
players_df_long = players_df.melt(id_vars=["Country"], var_name="Year", value_name="Number of Players")


df = pd.merge(fifa_ranking_df_long, players_df_long, on=["Country", "Year"], how="outer")
df.to_csv("df.csv", index=False) #download the merged dataset to check if it has been merged properly
df
```

## Step 4: Predictions and Analysis

Now that we have created a dataset, we split it into training and testing sets to see if we can predict GDP and GDNI using predictive analysis. 

First we train a linear regression model with the data after dropping all rows with missing values. We will split 80% of the data for the training set and 20% for the testing set. We will then use the mean squared error and the r squared metrics to evaluate the model. 

***Linear Regression to Predict GDP***

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

final_df = pd.read_csv("final_df.csv")

#Dropping rows with missing values
cleaned_df = final_df.dropna()

#Setting features and target variables
features = ['FIFA Ranking', 'Number of Players', 'Average Market Value (in Million USD)', 
            'Total Market Value (in Million USD)']
target = 'GDP'

X = cleaned_df[features]
y = cleaned_df[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=707)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

#Evaluating the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.4f}")
print(f"R-squared: {r2:.4f}")

#Plot
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7, label="Predicted vs Actual")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label="Ideal Fit")
plt.xlabel("Actual GDP")
plt.ylabel("Predicted GDP")
plt.title("Linear Regression: Actual vs Predicted GDP")
plt.legend()
plt.grid()
plt.show()
```

The mean squared error value and the r squared values indicate that the data we have found do not do well to predict GDP at all. We will try training a decision tree to see if we can capture more of the variance, and then a regression forest. 

***Decision Tree Algorithm to Predict GDP***

```python
# Predict GDP using the trained Decision Tree
from sklearn.tree import DecisionTreeRegressor

decision_tree_model = DecisionTreeRegressor(random_state=707)
decision_tree_model.fit(X_train, y_train)

y_pred_tree = decision_tree_model.predict(X_test)

mse_tree = mean_squared_error(y_test, y_pred_tree)
r2_tree = r2_score(y_test, y_pred_tree)

print(f"Mean Squared Error: {mse_tree:.4f}")
print(f"R-squared: {r2_tree:.4f}")

#Plot
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_tree, alpha=0.7, label="Predicted vs Actual GDP (Decision Tree)")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label="Ideal Fit")
plt.xlabel("Actual GDP")
plt.ylabel("Predicted GDP")
plt.title("Decision Tree: Actual vs Predicted GNI")
plt.legend()
plt.grid()
plt.show()
```

We found the a decision tree does better to capture more of the variance in GDP indicated by an r squared value of 0.1306. This indicates that footballing metrics we have found may account for about 13.06% of the variance in GDP. We now feed our data to a random forest algorithm.

***Random Forest Algorithm to Predict GDP***

```python
# Predict GDP using the trained Random Forest model
from sklearn.ensemble import RandomForestRegressor

random_forest_model = RandomForestRegressor(random_state=707, n_estimators=100)
random_forest_model.fit(X_train, y_train)

y_pred_rf = random_forest_model.predict(X_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print(f"Mean Squared Error: {mse_rf:.4f}")
print(f"R-squared: {r2_rf:.4f}")

#Finding Feature Importance for the trained Random Forest model
feature_importances = random_forest_model.feature_importances_
importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

print(importance_df)

#Visualizing feature importances

plt.figure(figsize=(10, 6))
plt.bar(importance_df['Feature'], importance_df['Importance'], alpha=0.8)
plt.title("Feature Importance from Random Forest")
plt.xlabel("Features")
plt.ylabel("Importance Score")
plt.tight_layout()
plt.show()
```

A random forest model was significantly better at predicting GDP being able to explain about 53.3% of the variance in GDP. While this is an improvement, we still have 46.75% of the variance that require more data to explain. 

To gain some insight into what kind of data we need, we extract the feature importance scores given by the random forest algorithm.

The importance scores illustrate that the most important feature was the 'number of players'. 'Number of players' was used in about 33% of decisions in the random forest algorithm. 

This feature represents the number of players from each country playing in the top 5 European leagues. This feature likely indicates the level of footballing infrastructure, i.e. the higher number means that the country possesses infrastracture capable of producing talents good enought to play in the top European leagues. 

Hence, we believe that more data on footballing infrastructure will help boost the model's predictive power. We also believe creating interaction terms between 'number of players' and 'total market value' may help uncover new relationships.

Moving forward with the project, collecting more data should be the main focus. Economic data on the infrastructure of football in countries will be helpful. We believe creating interaction terms between existing features will help boost prediction power as well. 

However, finding data on these subjects over a course of 20 years has proven difficult. Hence, we may want to bolster existing data as well. Using mean or median imputation to fill holes in incomplete observations that we already have. 

Another bit of change we can make to our data is to apply logarithmic transformations. GDP and GNI are huge numbers, applying a log transformation will help to reduce skewness, making relationships between features more linear (i.e. increase predictive power of the linear regression algorithm) and stabilize variance.
