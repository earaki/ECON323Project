```python
filtered_teams = [
    "England", "Spain", "Germany", "Brazil", "Italy", "Argentina", "Mexico",
    "USA", "France", "Saudi Arabia", "Turkey", "Qatar", "Korea Republic",
    "Indonesia", "Colombia", "Portugal", "Netherlands", "Poland", "Scotland",
    "Switzerland"
]

years = list(range(2003, 2023))

# Initialize an empty list to store the yearly DataFrames
dfs = []

# Loop through each year and read the corresponding cleaned FIFA ranking data
for year in years:
    file_path = f'Data sets/Fifa Rankings (2010-2024)/FIFA Ranking {year} Cleaned.csv'
    df = pd.read_csv(file_path)
    # Keep only the 'Team' and 'RK' columns
    df = df[['Team', 'RK']]
    # Filter by selected teams
    df = df[df['Team'].isin(filtered_teams)]
    # Rename the 'RK' column to indicate the year
    df = df.rename(columns={'RK': f'RK_{year}'})
    
    # Append the dataframe to the list
    dfs.append(df)

# Merge all DataFrames from the list
fifa_ranking_df = dfs[0]
for df in dfs[1:]:
    fifa_ranking_df = pd.merge(fifa_ranking_df, df, on='Team', how='outer')

# Load and merge 2023 rankings separately
fifa_ranking_2023 = pd.read_csv('Data sets/Fifa Rankings (2010-2024)/FIFA Ranking 2023 Cleaned.csv')
fifa_ranking_2023_filtered = fifa_ranking_2023[fifa_ranking_2023['Team'].isin(filtered_teams)]
fifa_ranking_2023_filtered = fifa_ranking_2023_filtered[['Team', 'RK']].rename(columns={'RK': 'RK_2023'})

fifa_ranking_df = pd.merge(fifa_ranking_df, fifa_ranking_2023_filtered, on='Team', how='left')

# Clean up column names and adjust formatting
fifa_ranking_df.columns = fifa_ranking_df.columns.str.replace('RK_', '')
fifa_ranking_df = fifa_ranking_df.rename(columns={'Team': 'Country'})
fifa_ranking_df["Country"] = fifa_ranking_df["Country"].replace("USA", "United States")  # Normalize country names
fifa_ranking_df.reset_index(drop=True, inplace=True)
fifa_ranking_df["Country"] = fifa_ranking_df["Country"].replace("Korea Republic", "South Korea")

fifa_ranking_df
```

```python
#List of countries to filter
filtered_teams = [
    "England", "Spain", "Germany", "Brazil", "Italy", "Argentina", "Mexico",
    "United States", "France", "Saudi Arabia", "Türkiye", "Qatar", "Korea Republic",
    "Indonesia", "Colombia", "Portugal", "Netherlands", "Poland", "Scotland",
    "Switzerland"]

#Load player data from multiple CSV files, filter by the given countries, and merge the results.
def load_and_filter_players_data(file_paths, filtered_teams):

    dfs = []
    
    for file_path in file_paths:
        df = pd.read_csv(file_path)
        df['Nation'] = df['Nation'].apply(lambda x: ' '.join(x.split()[1:]))
        if '# Players' in df.columns:
            df = df[df['Nation'].isin(filtered_teams)]
            year = file_path.split('/')[-1].split('-')[0]
            df = df[['Nation', '# Players']]
            df = df.rename(columns={'# Players': f'{year}'})
            dfs.append(df)

    if dfs:
        merged_df = dfs[0]
        for df in dfs[1:]:
            merged_df = pd.merge(merged_df, df, on='Nation', how='outer')

        merged_df.set_index('Nation', inplace=True)
        return merged_df
    else:
        print("No dataframes to merge.")
        return pd.DataFrame()

# Example usage with the uploaded files
file_paths = ["Data sets/NumberPlayers/2003-2004.csv", "Data sets/NumberPlayers/2004-2005.csv", 
              "Data sets/NumberPlayers/2005-2006.csv", "Data sets/NumberPlayers/2006-2007.csv", 
              "Data sets/NumberPlayers/2007-2008.csv", "Data sets/NumberPlayers/2008-2009.csv", 
              "Data sets/NumberPlayers/2009-2010.csv", "Data sets/NumberPlayers/2010-2011.csv", 
              "Data sets/NumberPlayers/2011-2012.csv", "Data sets/NumberPlayers/2012-2013.csv", 
              "Data sets/NumberPlayers/2013-2014.csv", "Data sets/NumberPlayers/2014-2015.csv",
              "Data sets/NumberPlayers/2015-2016.csv", "Data sets/NumberPlayers/2016-2017.csv",
              "Data sets/NumberPlayers/2017-2018.csv", "Data sets/NumberPlayers/2018-2019.csv", 
              "Data sets/NumberPlayers/2019-2020.csv", "Data sets/NumberPlayers/2020-2021.csv", 
              "Data sets/NumberPlayers/2021-2022.csv", "Data sets/NumberPlayers/2022-2023.csv", 
              "Data sets/NumberPlayers/2023-2024.csv"]
players_df = load_and_filter_players_data(file_paths, filtered_teams)
players_df.reset_index(inplace=True)
players_df = players_df.rename(columns = {'Nation':'Country'})
players_df["Country"] = players_df["Country"].replace("Türkiye", "Turkey") #ensure country names are same for merging
players_df["Country"] = players_df["Country"].replace("Korea Republic", "South Korea")
players_df.to_csv("players_df.csv", index=False)
players_df
```

```python
#Step 3: Merging the first two datasets into one
fifa_ranking_df_long = fifa_ranking_df.melt(id_vars=["Country"], var_name="Year", value_name="FIFA Ranking")
players_df_long = players_df.melt(id_vars=["Country"], var_name="Year", value_name="Number of Players")


df = pd.merge(fifa_ranking_df_long, players_df_long, on=["Country", "Year"], how="outer")
df.to_csv("df.csv", index=False) #download the merged dataset to check if it has been merged properly
df
```
